```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1 = [1,2,3,4,5,6]
data2 = [ 1,2,4,5,4,6]
x = 10*np.array(range(len(data1)))

plt.plot( x, data1, 'go') # Marcador verde
plt.plot( x, data1, 'k:', color='orange') # Linha pontilhada laranja

plt.plot( x, data2, 'r^') # Triangulo vermelho
plt.plot( x, data2, 'k--', color='blue')  # Linha tracejada azul

plt.axis([-10, 60, 0, 11])
plt.title("Evolução de recém nascido")

plt.grid(True)
plt.xlabel("Idade")
plt.ylabel("Semanas")
plt.show()
```
